```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

Stripplot

```python
_ = sns.catplot(tips, x='day', y='total_bill')  # catplot grupuje cała liste wykresów - domyslnie stripplot
```

Swarmplot

```python
_ = sns.catplot(tips, x='day', y='total_bill', kind='swarm')  # lepsze rozproszenie
```

Boxplot

```python
_ = sns.catplot(tips, x='day', y='total_bill',
                kind='box')  # kind - rodzaj wykresu, dół prostokątu 25% wartości, góra od 50% - 75% wartosci
```

Bar - pokazuje średnie

```python
_ = sns.catplot(tips, x='day', y='total_bill', kind='bar')
```

Countplot - określa ilość zliczeń

```python
_ = sns.catplot(tips, x='day', kind='count')
```
