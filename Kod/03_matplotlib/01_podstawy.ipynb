### Import biblioteki matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

Pierwszy wykres

```python
 _ = plt.plot([0, 1], [0, 1])  # _= usuwa nagłówek nad wykresem [<matplotlib.lines.Line2D at 0x2bec1c68bb0>]
```

```python
x = np.arange(-1, 1.1, 0.1)
y = x ** 2
```

```python
_ = plt.plot(x, y)
```

Dodać tytuł wykresu

```python
plt.plot(x, y)
_ = plt.title('Wykres x * x')
```

Dwie funkcje na jednym wykresie

```python
x1 = np.arange(-1, 1.1, 0.1)
y1 = x1 ** 3
```

```python
plt.plot(x, y)
_ = plt.plot(x1, y1)
```

Etykiety do wykresu

```python
plt.plot(x, y, label='x ^ 2')
plt.plot(x1, y1, label='x ^ 3')
_ = plt.legend()
```

Wiele wykresów na jednym obszarze

```python
plt.subplot(1, 2, 1)  # numeruje się od lewej do prawej( 1 wiersz, 2 kolumny, 1 komórka )
plt.plot(x, y)
plt.subplot(1, 2, 2)
_ = plt.plot(x1, y1)
```
